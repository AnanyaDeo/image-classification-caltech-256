```python
import numpy as np
import os
```

```python
dir_train = '/home/dan/data/Caltech-256/train/'
subdirs = list(os.walk(dir_train))[0][1]
```

```python
class_numbers = {int(x.split('.')[0]): x.split('.')[-1] for x in subdirs}
```

```python
for x in class_numbers:
    if '-101' in class_numbers[x]:
        class_numbers[x] = class_numbers[x][:-4]
```
